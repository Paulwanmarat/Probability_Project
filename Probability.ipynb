```python
import matplotlib.pyplot as plt
import numpy as np
from fractions import Fraction

genres = ["Pop", "Hip-hop", "Rock", "K-Pop", "Classical"]
students = [15, 2, 3, 5, 5]

total_students = sum(students)
print("Total Students:", total_students)
print("\nGenre Statistics")

for g, s in zip(genres, students):
    prob = s / total_students
    percent = prob * 100
    simplified_frac = Fraction(s, total_students)
    original_frac = f"{s}/{total_students}"
    
    print(f"{g}: {s} students | Fraction = {original_frac} = {simplified_frac} | Probability = {prob:.2f} | {percent:.1f}%")

print("\nExtra Statistics")
print("Mean:", np.mean(students))
print("Median:", np.median(students))
print("Standard Deviation:", round(np.std(students), 2))

ranking = sorted(zip(students, genres), reverse=True)
print("\nRanking (Most Popular to Least):")
for count, genre in ranking:
    print(f"{genre}: {count}")

#Bar chart hohoho
plt.figure(figsize=(10,6))
plt.bar(genres, students)
plt.title("Favorite Music Genre", fontsize=16, weight='bold')
plt.xlabel("Genre")
plt.ylabel("Number of Students")
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()


#Horizontal chart kub
plt.figure(figsize=(10,6))
plt.barh(genres, students)
plt.title("Favorite Music Genre (Horizontal)", fontsize=16, weight='bold')
plt.xlabel("Number of Students")
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.show()


#Pie Chart yim
plt.figure(figsize=(8,8))
wedges, texts, autotexts = plt.pie(
    students,
    labels=genres,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.6
)

for autotext in autotexts:
    autotext.set_fontsize(13)
    autotext.set_weight('bold')

plt.title("Favorite Music Genre", fontsize=16, weight='bold')
plt.axis('equal')
plt.show()

#Donut chart na kub
plt.figure(figsize=(8,8))

colors = ['#4C72B0', '#DD8452', '#55A868', '#C44E52', '#8172B3']

wedges, texts, autotexts = plt.pie(
    students,
    labels=genres,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    pctdistance=0.75,
    labeldistance=1.05,
    wedgeprops=dict(width=0.35, edgecolor='white', linewidth=2),
    colors=colors
)

plt.text(0, 0, "Music\nSurvey", ha='center', va='center',
         fontsize=16, weight='bold')

for text in texts:
    text.set_fontsize(12)

for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_weight('bold')

plt.title("Favorite Music Genre", fontsize=18, weight='bold', pad=20)
plt.axis('equal')
plt.tight_layout()
plt.show()


#Line chart na kub pom
plt.figure(figsize=(10,6))
plt.plot(genres, students, marker='o')
plt.title("Favorite Music Genre", fontsize=16, weight='bold')
plt.ylabel("Number of Students")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()


#Area Chart Yim
plt.figure(figsize=(10,6))
plt.fill_between(genres, students)
plt.title("Favorite Music Genre", fontsize=16, weight='bold')
plt.ylabel("Number of Students")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()

```
